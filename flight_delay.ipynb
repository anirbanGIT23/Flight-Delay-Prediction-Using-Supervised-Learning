```python
from sklearn.feature_selection import SelectPercentile, f_regression
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import pandas as pd

df_encoded = df_seen.copy()

# Encode categorical features
for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col].astype(str))

# Drop rows with missing target
df_clean = df_encoded.dropna(subset=['ARRIVAL_DELAY'])

# Define known post-departure / leaky features
leaky_features = [
    'DEP_TIME', 'DEP_DELAY', 'TAXI_OUT', 'WHEELS_OFF',
    'TAXI_IN', 'ARR_DELAY', 'ARRIVAL_DELAY',
    'CANCELLED', 'CANCELLATION_CODE',
    'DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT', 'WHEELS_ON', 'ACTUAL_ELAPSED_TIME'
]

# Drop these from the dataset if they exist
X = df_clean.drop(columns=[col for col in leaky_features if col in df_clean.columns], errors='ignore')

# Target
y_reg = df_clean['ARRIVAL_DELAY']

# Impute missing values
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Sample if needed (optional for performance)
sample_size = min(250000, len(X_imputed))  # NOTE: changed to 250k to match your real dataset
X_sample = X_imputed.sample(sample_size, random_state=42)
y_sample = y_reg.loc[X_sample.index]

# Select top 50% of features using F-test (only from clean features)
selector = SelectPercentile(score_func=f_regression, percentile=50)
selector.fit(X_sample, y_sample)
selected_mask = selector.get_support()
selected_features_reg = X.columns[selected_mask]

# Output
reg_output = pd.DataFrame({'Selected_Features_Regression': selected_features_reg})
reg_output.loc[len(reg_output.index)] = ['Total Selected: ' + str(len(selected_features_reg))]

print("\n✅ Selected features for predicting delay duration (in minutes):")
print(reg_output)
```

```python
# probability of delay in percentage -> uses anova for classification i.e binary outputs

from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import pandas as pd

df_encoded = df_seen.copy()

# Encode categorical columns
for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col].astype(str))

# Create classification target
df_encoded['DELAY_FLAG'] = (df_encoded['ARRIVAL_DELAY'] > 15).astype(int)

# Drop rows with missing target
df_clean = df_encoded.dropna(subset=['DELAY_FLAG'])

# **Exclude the delay component columns here to avoid leakage**
delay_components = ['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT']

X = df_clean.drop(columns=['ARRIVAL_DELAY', 'DELAY_FLAG'] + delay_components)  # Excluded delay components
y_cls = df_clean['DELAY_FLAG']

# Impute missing values
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Use the entire df_seen for feature selection (sample size = all rows)
sample_size = min(len(X_imputed), 2500000)
X_sample = X_imputed.sample(sample_size, random_state=42)
y_sample = y_cls.loc[X_sample.index]

# Feature selection with f_classif for classification
selector = SelectPercentile(score_func=f_classif, percentile=50)
selector.fit(X_sample, y_sample)
selected_mask = selector.get_support()
selected_features_cls = X.columns[selected_mask]

cls_output = pd.DataFrame({
    'Selected_Features_Classification': selected_features_cls
})
cls_output.loc[len(cls_output.index)] = ['Total Selected: ' + str(len(selected_features_cls))]

print("\nSelected features for predicting delay probability (classification):")
print(cls_output)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectPercentile, f_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
df_seen = df_shuffled.iloc[:2500000].copy()
df_unseen = df_shuffled.iloc[2500000:].copy()

df_seen_encoded = df_seen.copy()
for col in df_seen_encoded.select_dtypes(include='object').columns:
    df_seen_encoded[col] = LabelEncoder().fit_transform(df_seen_encoded[col].astype(str))

df_seen_clean = df_seen_encoded.dropna(subset=['ARRIVAL_DELAY'])

X_seen = df_seen_clean.drop(columns=['ARRIVAL_DELAY'])
y_seen = df_seen_clean['ARRIVAL_DELAY']

imputer = SimpleImputer(strategy='median')
X_seen_imputed = pd.DataFrame(imputer.fit_transform(X_seen), columns=X_seen.columns)

selector = SelectPercentile(score_func=f_regression, percentile=50)
selector.fit(X_seen_imputed, y_seen)
selected_mask = selector.get_support()
selected_features_reg = X_seen.columns[selected_mask]

X_reg = X_seen_imputed[selected_features_reg]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_seen, test_size=0.2, random_state=42
)

reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)

y_pred = reg_model.predict(X_test_reg)
print("Train-Test Split Evaluation:")
print("MAE: ", round(mean_absolute_error(y_test_reg, y_pred), 2))
print("MSE: ", round(mean_squared_error(y_test_reg, y_pred), 2))
print("R² Score:", round(r2_score(y_test_reg, y_pred), 4))

df_unseen_encoded = df_unseen.copy()
for col in df_unseen_encoded.select_dtypes(include='object').columns:
    df_unseen_encoded[col] = LabelEncoder().fit_transform(df_unseen_encoded[col].astype(str))

df_unseen_clean = df_unseen_encoded.dropna(subset=['ARRIVAL_DELAY'])
y_unseen = df_unseen_clean['ARRIVAL_DELAY']
X_unseen = df_unseen_clean.drop(columns=['ARRIVAL_DELAY'])

X_unseen_imputed = pd.DataFrame(imputer.transform(X_unseen), columns=X_unseen.columns)
X_unseen_final = X_unseen_imputed[selected_features_reg]

y_unseen_pred = reg_model.predict(X_unseen_final)
print("\nUnseen Data Evaluation:")
print("MAE: ", round(mean_absolute_error(y_unseen, y_unseen_pred), 2))
print("MSE: ", round(mean_squared_error(y_unseen, y_unseen_pred), 2))
print("R² Score:", round(r2_score(y_unseen, y_unseen_pred), 4))
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import pandas as pd

df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
df_seen = df_shuffled.iloc[:2500000].copy()
df_unseen = df_shuffled.iloc[2500000:].copy()

df_seen_encoded = df_seen.copy()
for col in df_seen_encoded.select_dtypes(include='object').columns:
    df_seen_encoded[col] = LabelEncoder().fit_transform(df_seen_encoded[col].astype(str))

df_seen_encoded['DELAY_CLASS'] = (df_seen_encoded['ARRIVAL_DELAY'] > 15).astype(int)
df_seen_clean = df_seen_encoded.dropna(subset=['ARRIVAL_DELAY'])

X_seen = df_seen_clean.drop(columns=['ARRIVAL_DELAY', 'DELAY_CLASS'])
y_cls = df_seen_clean['DELAY_CLASS']

imputer = SimpleImputer(strategy='median')
X_seen_imputed = pd.DataFrame(imputer.fit_transform(X_seen), columns=X_seen.columns)

selector_cls = SelectPercentile(score_func=f_classif, percentile=50)
selector_cls.fit(X_seen_imputed, y_cls)
selected_mask_cls = selector_cls.get_support()
selected_features_cls = X_seen.columns[selected_mask_cls]

X_cls = X_seen_imputed[selected_features_cls]


X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X_cls, y_cls, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_cls_scaled = scaler.fit_transform(X_train_cls)
X_test_cls_scaled = scaler.transform(X_test_cls)

cls_model = LogisticRegression(max_iter=2000, solver='lbfgs')
cls_model.fit(X_train_cls_scaled, y_train_cls)

y_pred_cls = cls_model.predict(X_test_cls_scaled)

print("Confusion Matrix:")
print(confusion_matrix(y_test_cls, y_pred_cls))

print("\nClassification Report:")
print(classification_report(y_test_cls, y_pred_cls))

print(f"Accuracy: {accuracy_score(y_test_cls, y_pred_cls):.4f}")


```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE

# Create binary target
sampled_df['DELAY_BINARY'] = sampled_df['DELAY_BUCKET'].apply(lambda x: 0 if x == 'On Time' else 1)

# Features to use
features_to_use = [
    'CRS_DEP_TIME', 'CRS_ARR_TIME', 'CRS_ELAPSED_TIME', 'DISTANCE',
    'FL_DATE', 'ORIGIN', 'DEST', 'AIRLINE'
]

X = sampled_df[features_to_use].copy()
y = sampled_df['DELAY_BINARY']

# Feature engineering
X['CRS_DEP_HOUR'] = X['CRS_DEP_TIME'] // 100
X['CRS_ARR_HOUR'] = X['CRS_ARR_TIME'] // 100
X['FL_DATE'] = pd.to_datetime(X['FL_DATE'])
X['DAY_OF_WEEK'] = X['FL_DATE'].dt.dayofweek
X['MONTH'] = X['FL_DATE'].dt.month
X = X.drop(columns=['CRS_DEP_TIME', 'CRS_ARR_TIME', 'FL_DATE'])

# Define categorical and numeric columns
categorical_cols = ['ORIGIN', 'DEST', 'AIRLINE', 'DAY_OF_WEEK', 'MONTH']
numeric_cols = ['CRS_ELAPSED_TIME', 'DISTANCE', 'CRS_DEP_HOUR', 'CRS_ARR_HOUR']

# Preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
    ]
)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# Preprocess training and test data
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_processed, y_train)

# Train model
rf_model = RandomForestClassifier(random_state=42, n_jobs=-1)
rf_model.fit(X_resampled, y_resampled)

# Predict and evaluate
y_pred = rf_model.predict(X_test_processed)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Feature importance
ohe = preprocessor.named_transformers_['cat']
ohe_feature_names = ohe.get_feature_names_out(categorical_cols)
all_feature_names = numeric_cols + list(ohe_feature_names)
importances = rf_model.feature_importances_

feature_importance_df = pd.DataFrame({
    'Feature': all_feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("\nTop features:\n", feature_importance_df.head(20))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# Create binary target
sampled_df['DELAY_BINARY'] = sampled_df['DELAY_BUCKET'].apply(lambda x: 0 if x == 'On Time' else 1)

# Features to use (pre-departure only)
features_to_use = [
    'CRS_DEP_TIME', 'CRS_ARR_TIME', 'CRS_ELAPSED_TIME', 'DISTANCE',
    'FL_DATE', 'ORIGIN', 'DEST', 'AIRLINE'
]

X = sampled_df[features_to_use].copy()
y = sampled_df['DELAY_BINARY']

# Feature engineering
X['CRS_DEP_HOUR'] = X['CRS_DEP_TIME'] // 100
X['CRS_ARR_HOUR'] = X['CRS_ARR_TIME'] // 100
X['FL_DATE'] = pd.to_datetime(X['FL_DATE'])
X['DAY_OF_WEEK'] = X['FL_DATE'].dt.dayofweek
X['MONTH'] = X['FL_DATE'].dt.month
X = X.drop(columns=['CRS_DEP_TIME', 'CRS_ARR_TIME', 'FL_DATE'])

# Define categorical and numerical columns
categorical_cols = ['ORIGIN', 'DEST', 'AIRLINE', 'DAY_OF_WEEK', 'MONTH']
numeric_cols = ['CRS_ELAPSED_TIME', 'DISTANCE', 'CRS_DEP_HOUR', 'CRS_ARR_HOUR']

# Preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ]
)

# Full pipeline with SMOTE + LightGBM
clf = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('classifier', LGBMClassifier(random_state=42, n_jobs=-1))
])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# Train
clf.fit(X_train, y_train)

# Predict & Evaluate
y_pred = clf.predict(X_test)
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Feature importance (optional)
importances = clf.named_steps['classifier'].feature_importances_
ohe = clf.named_steps['preprocessor'].named_transformers_['cat']
ohe_feature_names = ohe.get_feature_names_out(categorical_cols)
all_feature_names = numeric_cols + list(ohe_feature_names)
feature_importance_df = pd.DataFrame({
    'Feature': all_feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print("\nTop features:\n", feature_importance_df.head(20))
```
